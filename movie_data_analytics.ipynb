```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) #Adjusting the configuration of the plots I am using

# Reading in the data

df = pd.read_csv(r'C:\Users\imman\Downloads\moviedata\movies.csv')
df = df.dropna()
```

```python
#Looking at data to see if everything was imported correctly
df.head(10)
```

```python
#Checking for missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# Column Data types
df.dtypes
```

```python
#Removing unnecessary decimal places by changing data type to int

df['budget'] = df['budget'].fillna(0).astype('int64')
df['votes'] = df['votes'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df
```

```python
#Correcting the year column to match the release date year

df['Correct Year'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
#Sorting movies by how much they made

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
#Scatter plot with budget vs gross revenue
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Film Budget')
plt.show()
```

```python
#Plot Budget vs Gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color': 'blue'})
```

```python
#Looking at the numerical correlation using pearson method

df.corr()
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Converting the object values to unique numerical values to get their correlation
df_conversion = df

for col_name in df_conversion.columns:
    if df_conversion[col_name].dtype == 'object':
        df_conversion[col_name] = df_conversion[col_name].astype('category')
        df_conversion[col_name] = df_conversion[col_name].cat.codes

df_conversion

```

```python
df
```

```python
#Making a correlation matrix for all values

correlation_matrix = df_conversion.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_conversion.corr()
```

```python
#Improving data readability by unstacking it

correlation_mat = df_conversion.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# Improving data readability by sorting

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#Looking at highly correlated data
high_correlation = sorted_pairs[(sorted_pairs) > .5]
high_correlation
```
